```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("bunga.jpg")
```

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

```python
img_biner = np.zeros(img.shape, dtype=np.uint8)
```

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue))/3
        if gray < 128:
            gray = 0
        else:
            gray = 255
        img_biner[y][x] = (gray, gray, gray)
```

```python
plt.imshow(img_biner)
plt.title("Biner")
plt.show()
```
