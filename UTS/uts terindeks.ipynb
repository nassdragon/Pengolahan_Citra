```python
import numpy as np
from PIL import Image
```

```python
img = Image.open("bunga.jpg")
img = img.convert('RGB')
width, height = img.size
```

```python
palette = {
    0: (0, 0, 0),
    2: (255, 0, 0),
    3: (0, 255, 0),
    4: (0, 0, 255),
}
```

```python
import math

def find_closest_color_index(pixel):
    closest_index = None
    min_distance = float('inf')

    for index in palette:
        r1, g1, b1 = pixel
        r2, g2, b2 = palette[index]
        distance = math.sqrt((r1 - r2)**2 + (g1 - g2)**2 + (b1 - b2)**2)

        if distance < min_distance:
            min_distance = distance
            closest_index = index

    return closest_index
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.title('Gambar Asli')
plt.show()
```

```python
matrix = []

for y in range(height):
    row = []
    for x in range(width):
        pixel = img.getpixel((x, y))
        index = find_closest_color_index(pixel)
        row.append(index)
    matrix.append(row)
```

```python
img_terindeks = Image.new('RGB', (width, height))

for y in range(height):
    for x in range(width):
        color = palette[matrix[y][x]]
        img_terindeks.putpixel((x, y), color)
```

```python
plt.imshow(img_terindeks)
plt.title('Gambar Setelah di Konversi')
plt.show()
```
